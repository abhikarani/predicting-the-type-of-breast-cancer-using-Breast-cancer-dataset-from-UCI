```python
import os
import math
import seaborn as sns
import matplotlib.pyplot as plt
from termcolor import colored



from sklearn.model_selection import RandomizedSearchCV, KFold,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor,StackingRegressor, RandomForestRegressor, ExtraTreesRegressor
from sklearn.linear_model import Lasso
from sklearn.pipeline import make_pipeline

from scipy.stats import skew
from scipy.special import boxcox1p
from scipy.stats import boxcox_normmax
from scipy.stats import boxcox
from scipy.special import inv_boxcox

from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler, OneHotEncoder
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Breast_cancer_dataset_from_UCI.csv")
```

```python
data.head()
```

```python
data['diagnosis']=data['diagnosis'].astype('category')
data['diagnosis_new']=data['diagnosis'].cat.codes
enc=OneHotEncoder()
enc_data=pd.DataFrame(enc.fit_transform(data[['diagnosis_new']]).toarray())
data=data.merge(enc_data, right_index = True, left_index=True)
```

```python
data.head()
data.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
corr = data.corr()
corr
```

```python
corr_dict = corr["diagnosis_new"].sort_values(ascending=False).to_dict()
corr_dict
```

```python
#from this I can see which parameters affect the diagnosis type the most
pos_corr = {k for k,v in corr_dict.items() if v > 0.1 and v != 1} 
pos_corr
```

```python
#print the most negatively correlated features
neg_corr = {k:v for k,v in corr_dict.items() if v < 0 and v > -1}
neg_corr
```

```python
data.head()
```

```python
X_old = data.drop(data.iloc[:, 32:36], axis=1)
```

```python
X = X_old.drop(X_old.iloc[:,1:2], axis=1)
y = data["diagnosis_new"]
```

```python
X.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

```python
X_train_feature_selection = X_train[pos_corr]
X_valid_feature_selection = X_valid[pos_corr]
```

```python
X_train_feature_selection.head()
```

```python
#Bayesian Classifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
y_train.head()
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train) #without feature selection
y_pred = gnb.predict(X_valid)
print("using Bayesian classifier without feature selection")
print("Accuracy:",metrics.accuracy_score(y_valid, y_pred))
print("Precision:",metrics.precision_score(y_valid, y_pred))
print("Recall:",metrics.recall_score(y_valid, y_pred))
print("F1-score:",metrics.f1_score(y_valid, y_pred))
print("Error rate:",1-metrics.accuracy_score(y_valid, y_pred))
```

```python
gnb.fit(X_train_feature_selection, y_train) #with feature selection
y_pred = gnb.predict(X_valid_feature_selection)
print("using Bayesian classifier with feature selection")
print("Accuracy:",metrics.accuracy_score(y_valid, y_pred))
print("Precision:",metrics.precision_score(y_valid, y_pred))
print("Recall:",metrics.recall_score(y_valid, y_pred))
print("F1-score:",metrics.f1_score(y_valid, y_pred))
print("Error rate:",1-metrics.accuracy_score(y_valid, y_pred))
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_valid)
print("using decision tree without feature selection")
print("Accuracy:",metrics.accuracy_score(y_valid, y_pred))
print("Precision:",metrics.precision_score(y_valid, y_pred))
print("Recall:",metrics.recall_score(y_valid, y_pred))
print("F1-score:",metrics.f1_score(y_valid, y_pred))
print("Error rate:",1-metrics.accuracy_score(y_valid, y_pred))
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train_feature_selection,y_train)
y_pred = dtc.predict(X_valid_feature_selection)
print("using decision tree with feature selection")
print("Accuracy:",metrics.accuracy_score(y_valid, y_pred))
print("Precision:",metrics.precision_score(y_valid, y_pred))
print("Recall:",metrics.recall_score(y_valid, y_pred))
print("F1-score:",metrics.f1_score(y_valid, y_pred))
print("Error rate:",1-metrics.accuracy_score(y_valid, y_pred))
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_valid)
print("using random forest classifier without feature selection")
print("Accuracy:",metrics.accuracy_score(y_valid, y_pred))
print("Precision:",metrics.precision_score(y_valid, y_pred))
print("Recall:",metrics.recall_score(y_valid, y_pred))
print("F1-score:",metrics.f1_score(y_valid, y_pred))
print("Error rate:",1-metrics.accuracy_score(y_valid, y_pred))
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train_feature_selection,y_train)
y_pred=rfc.predict(X_valid_feature_selection)
print("using random forest classifier with feature selection")
print("Accuracy:",metrics.accuracy_score(y_valid, y_pred))
print("Precision:",metrics.precision_score(y_valid, y_pred))
print("Recall:",metrics.recall_score(y_valid, y_pred))
print("F1-score:",metrics.f1_score(y_valid, y_pred))
print("Error rate:",1-metrics.accuracy_score(y_valid, y_pred))
```
